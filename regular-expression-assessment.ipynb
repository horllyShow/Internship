Question 1- Write a Python program to replace all occurrences of a space, comma, or dot with a colon.
Sample Text- 'Python Exercises, PHP exercises.'
Expected Output: Python:Exercises::PHP:exercises:


```python
import re

def replace_with_colon(text):
    # Use regular expression to replace space, comma, and dot with a colon
    pattern = r'[ ,.]'
    result = re.sub(pattern, ':', text)
    return result

sample_text = 'Python Exercises, PHP exercises.'
output_text = replace_with_colon(sample_text)
print(output_text)
```

Question 3- Create a function in python to find all words that are at least 4 characters long in a string. The use of the re.compile() method is mandatory.

```python


def find_long_words(text):
    # Define a regular expression pattern to match words with at least 4 characters
    pattern = re.compile(r'\b\w{4,}\b')

    # Use the findall() method to find all matching words in the text
    long_words = pattern.findall(text)

    return long_words

# Example usage:
text = "This is a sample text with some words of different lengths like apple, pear, and banana."
result = find_long_words(text)
print(result)
```

Question 4- Create a function in python to find all three, four, and five character words in a string. The use of the re.compile() method is mandatory.

```python


def find_words_by_length(text):
    # Define a regular expression pattern to match words with three, four, or five characters
    pattern = re.compile(r'\b\w{3,5}\b')

    # Use the findall() method to find all matching words in the text
    words = pattern.findall(text)

    return words

# Example usage:
text = "This is a sample text with some words of different lengths like apple, pear, and banana."
result = find_words_by_length(text)
print(result)
```

Question 5- Create a function in Python to remove the parenthesis in a list of strings. The use of the re.compile() method is mandatory.
Sample Text: ["example (.com)", "hr@fliprobo (.com)", "github (.com)", "Hello (Data Science World)", "Data (Scientist)"]
Expected Output:
example.com
hr@fliprobo.com
github.com
Hello Data Science World
Data Scientist

```python


def remove_parentheses(strings):
    # Define a regular expression pattern to match parentheses
    pattern = re.compile(r'\(([^)]*)\)')

    # Use the sub() method to remove the parentheses while keeping the contents
    result = [pattern.sub(r'\1', string) for string in strings]

    return result

# Example usage:
sample_text = ["example (.com)", "hr@fliprobo (.com)", "github (.com)", "Hello (Data Science World)", "Data (Scientist)"]
output = remove_parentheses(sample_text)
for item in output:
    print(item)
```

Question 6- Write a python program to remove the parenthesis area from the text stored in the text file using Regular Expression.

Expected Output: ["example", "hr@fliprobo", "github", "Hello", "Data"]
Note- Store given sample text in the text file and then to remove the parenthesis area from the text.


```python


# Function to remove the parentheses area using regular expressions
def remove_parentheses_area(text):
    pattern = re.compile(r'\([^)]*\)')
    return pattern.sub('', text)

# Read the content from the text file
file_path = "/kaggle/input/sample22/sample22.txt"

with open(file_path, "r") as file:
    lines = file.readlines()

# Process the lines to remove parentheses area
processed_lines = [remove_parentheses_area(line) for line in lines]

# Write the modified content back to the text file
new_file_path = "/kaggle/working/new_sample_data.txt"
with open(new_file_path, "w") as file:
    file.writelines(processed_lines)

# Print the modified content
for line in processed_lines:
    print(line.strip())
```

Question 9- Create a function in python to insert spaces between words starting with capital letters or with numbers.
Sample Text: “RegularExpression1IsAn2ImportantTopic3InPython"
Expected Output:  RegularExpression 1 IsAn 2 ImportantTopic 3 InPython

```python


def insert_spaces_before_caps_or_numbers(text):
    # Use regular expression to insert spaces between words starting with capital letters or numbers
    result = re.sub(r'([A-Z0-9][a-z]+)', r' \1', text)
    result = re.sub(r'([a-z])([0-9])', r'\1 \2', result)
    return result

# Example usage:
sample_text = "RegularExpression1IsAn2ImportantTopic3InPython"
output_text = insert_spaces_before_caps_or_numbers(sample_text)
print(output_text)
```

Question 11- Write a Python program to match a string that contains only upper and lowercase letters, numbers, and underscores.

```python

def is_valid_string(input_string):
    # Define a regular expression pattern to match valid strings
    pattern = r'^[a-zA-Z0-9_]+$'

    # Use re.match() to check if the input string matches the pattern
    if re.match(pattern, input_string):
        return True
    else:
        return False

# Example usage:
strings_to_check = ["Hello_World123"]

for string in strings_to_check:
    if is_valid_string(string):
        print(f"'{string}' is a valid string.")
    else:
        print(f"'{string}' is not a valid string.")
```

Question 12- Write a Python program where a string will start with a specific number. 

```python

def starts_with_number(input_string, number):
    # Create a regular expression pattern to match the specific number at the beginning of the string
    pattern = re.compile(r'^' + str(number) + r'.*')

    # Use re.match() to check if the input string matches the pattern
    if pattern.match(input_string):
        return True
    else:
        return False

# Example usage:
input_string = "123abc"
specified_number = 123

if starts_with_number(input_string, specified_number):
    print(f"The string '{input_string}' starts with the number {specified_number}.")
else:
    print(f"The string '{input_string}' does not start with the number {specified_number}.")
```

Question 13- Write a Python program to remove leading zeros from an IP address

```python

def remove_leading_zeros(ip_address):
    # Define a regular expression pattern to match and remove leading zeros
    pattern = re.sub(r'\b0+(\d)', r'\1', ip_address)

    return pattern

# Example usage:
ip_address = "192.001.001.001"
result = remove_leading_zeros(ip_address)
print(result)
```

Question 14- Write a regular expression in python to match a date string in the form of Month name followed by day number and year stored in a text file.
Sample text :  ' On August 15th 1947 that India was declared independent from British colonialism, and the reins of control were handed over to the leaders of the Country’.
Expected Output- August 15th 1947
Note- Store given sample text in the text file and then extract the date string asked format.


```python
# Read the content from the text file
file_path = r"/kaggle/input/sample-independence/sample22.txt"

with open(file_path, "r") as file:
    text = file.read()

# Define a regular expression pattern to match the desired date format
pattern = r'\b[A-Z][a-z]+\s\d+(?:st|nd|rd|th)\s\d{4}\b'

# Use re.search() to find the date in the specified format
match = re.search(pattern, text)

if match:
    date_string = match.group(0)
    print("Extracted date:", date_string)
else:
    print("Date not found in the specified format.")
```

Question 15- Write a Python program to search some literals strings in a string. 
Sample text : 'The quick brown fox jumps over the lazy dog.'
Searched words : 'fox', 'dog', 'horse'


```python

sample_text = 'The quick brown fox jumps over the lazy dog.'
searched_words = ['fox', 'dog', 'horse']

for word in searched_words:
    pattern = re.compile(r'\b' + re.escape(word) + r'\b')
    if pattern.search(sample_text):
        print(f"'{word}' found in the text.")
    else:
        print(f"'{word}' not found in the text.")
```

Question 16- Write a Python program to search a literals string in a string and also find the location within the original string where the pattern occurs
Sample text : 'The quick brown fox jumps over the lazy dog.'
Searched words : 'fox'


```python
sample_text = 'The quick brown fox jumps over the lazy dog.'
searched_word = 'fox'

pattern = re.compile(re.escape(searched_word))

match = pattern.search(sample_text)

if match:
    start = match.start()
    end = match.end()
    print(f"'{searched_word}' found in the text at position {start}-{end}.")
else:
    print(f"'{searched_word}' not found in the text.")
```

Question 18- Write a Python program to find the occurrence and position of the substrings within a string.

```python
sample_text = 'Python exercises, PHP exercises, C# exercises'
pattern = 'exercises'

# Use re.finditer to find all occurrences of the pattern and their positions
matches = re.finditer(pattern, sample_text)

# Iterate through the matches and print their position and occurrence
for match in matches:
    start = match.start()
    end = match.end()
    occurrence = sample_text[start:end]
    position = start
    print(f"Occurrence: {occurrence}, Position: {position}")
```

Question 20- Create a function in python to find all decimal numbers with a precision of 1 or 2 in a string. The use of the re.compile() method is mandatory.
Sample Text: "01.12 0132.123 2.31875 145.8 3.01 27.25 0.25"
Expected Output: ['01.12', '145.8', '3.01', '27.25', '0.25']

```python
def find_decimal_numbers(text):
    # Define a regular expression pattern with a precision of 1 or 2 decimal places
    pattern = re.compile(r'\b\d+\.\d{1,2}\b')

    # Use the findall method to find all matching numbers
    matches = pattern.findall(text)

    return matches

# Sample text
sample_text = "01.12 0132.123 2.31875 145.8 3.01 27.25 0.25"

# Call the function to find decimal numbers
result = find_decimal_numbers(sample_text)

# Print the result
print(result)
```
